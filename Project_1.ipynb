```python
import astropy.constants as const # import a package to call value of certain constants.
import astropy.units as u # import a package to call certain units.
import numpy as np # import a package to call math functions.
```

# Problem 1
## part a

```python
R_sun_earth = const.au # generate the distance between sun and earth
sun_mass = const.M_sun # generate the mass of sun
G = const.G # generate the gravitational constant
v_orbital = np.sqrt(G * sun_mass / R_sun_earth) # find the orbitasl velocity of earth around the Sun
print(v_orbital.to(u.km / u.s)) # convert the velocity to km/s and print the velocity
```

The orbital velocity of Earth around the Sun is 29.784691829676934 km/s.





```python
earth_mass = const.M_earth # generate the mass of earth
satellite_distance = (const.R_earth + 400 * u.km)  # generate the distance between earth and the satellite
v_orbital_satellite = np.sqrt(G * earth_mass / satellite_distance) # find the orbital velocity of the satellite around earth
print(v_orbital_satellite.to(u.km / u.s)) # convert the velocity to km/s and print the velocity
```

The orbital velocity of the satellite around Earth is 7.6685787036856965 km/s.

## part b

```python
blackhole_mass = const.M_sun # generate the mass of the blackhole
schwarzschild_radius = 2 * G * blackhole_mass / const.c**2  # find the radius of the blackhole
print(schwarzschild_radius.to(u.km)) # convert the radius to km and print the radius and print the radius
```


The Schwarzschild radius of a solar mass blackhole is 2.95325007610025 km.

# Problem 2

```python
import matplotlib.pyplot as plt # import a package to call plot functions.
blackhole_mass = 1e8 * const.M_sun # Define the mass of the black hole
radii = np.logspace(0, 3, 100) * u.kpc # Define the range of radii to calculate the rotation curve for
v_orbital = np.sqrt(G * blackhole_mass / radii) # Calculate the orbital velocity at each radius
plt.plot(radii.value, v_orbital.to(u.km / u.s).value) # Plot the rotation curve
plt.xscale('log') # Set the x-axis to logarithmic scale
plt.xlabel('Radius (kpc)')  # Set the x-axis label
plt.ylabel('Orbital Velocity (km/s)') # Set the y-axis label
plt.title('Rotation Curve of a Galaxy with a Central Black Hole') # Set the title of the plot
plt.show() # Show the plot
```
